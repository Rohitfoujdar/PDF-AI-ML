```python
import pandas as pd
```

```python
dataset = pd.DataFrame({"Size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
dataset.head(3)
```

```python
ord_data = [["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(dataset[["Size"]])
```

```python
dataset["size_en"]=oe.transform(dataset[["Size"]])
```

```python
dataset
```

```python
ord_data1 ={"s":9 , "m":8 , "l":7 , "xl":6} #create a dictionary
```

```python
dataset["size_en_map3"] = dataset["Size"].map(ord_data1) #Use the map function to map the size to the corresponding ordinal value
```
