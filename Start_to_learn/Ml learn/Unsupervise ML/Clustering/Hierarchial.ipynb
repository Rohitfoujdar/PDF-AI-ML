```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
# Create synthetic data
X, y = make_blobs(n_samples=100, centers=4, cluster_std=1.2, random_state=42)
```

```python
# Step 1: Plot Dendrogram using scipy
linked = linkage(X, method='ward')  # Ward linkage minimizes variance within clusters
```

```python
plt.figure(figsize=(12, 6))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title("📊 Dendrogram (Hierarchical Clustering)")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.grid(True)
plt.show()
```

```python
# Step 2: Apply Agglomerative Clustering to assign cluster labels
model = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
labels = model.fit_predict(X)

# Step 3: Plot clustered data points
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels, palette="tab10", s=80)
plt.title("🔗 Hierarchical Clustering (Agglomerative) Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend(title="Cluster")
plt.grid(True)
plt.show()
```
