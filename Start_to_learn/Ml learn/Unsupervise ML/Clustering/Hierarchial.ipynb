```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
# Create synthetic data
X, y = make_blobs(n_samples=100, centers=4, cluster_std=1.2, random_state=42)
```

```python
# Step 1: Plot Dendrogram using scipy
linked = linkage(X, method='ward')  # Ward linkage minimizes variance within clusters
```

```python
plt.figure(figsize=(12, 6))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title("ðŸ“Š Dendrogram (Hierarchical Clustering)")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
```

```python
dataset = pd.read_csv("iris_dataset.csv")
dataset.head()
x = dataset.iloc[:,:-1]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(x)
```

```python
scaled
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca_scaled = pca.fit_transform(scaled)
```

```python
plt.scatter(pca_scaled[:,0], pca_scaled[:,1])
```

```python
pca_scaled.shape
```

```python
#Agglomerative clustering
import scipy.cluster.hierarchy as sc
```

```python
plt.Figure(figsize=(20,7))
plt.title("Dendogram")
```

```python
#create deendogram
sc.dendrogram(sc.linkage(pca_scaled, method="ward"))
plt.title("Dendogram")
plt.xlabel("Sample Index")
plt.ylabel("Eucledian Distance")
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster = AgglomerativeClustering(n_clusters=2 , linkage="ward")
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0], pca_scaled[:,1], c=cluster.labels_)
```
