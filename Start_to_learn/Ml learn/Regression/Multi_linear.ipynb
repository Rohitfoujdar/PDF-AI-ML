```python
x = dataset.iloc[:,:-1] # features
y= dataset["Salary"]
x , y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test , y_train , y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
# y=m1x1+m2x2....+c
```

```python
lr.score(x_test, y_test)*100 #what is the accuracy of the model
```

```python
lr.coef_ # find the value of M1, M2, and M3
```

```python
lr.intercept_ # find the intercept of the regression line
```

```python
import pandas as pd
import numpy as np

# Random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
levels = np.random.randint(1, 21, 500)  # Job levels from 1 to 20
salaries = 30000 + (levels ** 2) * 250 + np.random.normal(0, 5000, 500)  # Non-linear relation with some noise

# Create DataFrame
df = pd.DataFrame({
    "Level": levels,
    "Salary": salaries.astype(int)
})

# Save to CSV
file_path = "level_salary_data.csv"
df.to_csv(file_path, index=False)

file_path
```
