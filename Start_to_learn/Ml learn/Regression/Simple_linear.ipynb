```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("CGPA_Package_Data.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
x= dataset[["CGPA"]]
y= dataset["PACKAGE"]
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="CGPA" , y="PACKAGE" , data=dataset)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 ,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression # Import Linear Regression
```

```python
lr = LinearRegression() # initialize linear regression model
```

```python
lr.fit(x_train, y_train)
```

```python
lr.predict([[6.87]]) # Predict ing the value of the house
```

```python
lr.score(x_test, y_test)*100  # calculate accuracy
```

```python
# y=m*x+c
```

```python
lr.coef_   # Value of the coefficient (m)
```

```python
lr.intercept_ # Value of intercept (c)
```

```python
2.54662209*6.87-0.35586978840211003
```

```python
y_prd=lr.predict(x)
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x="CGPA" , y="PACKAGE" , data=dataset)
plt.plot(dataset["CGPA"], y_prd , c="red")
plt.legend(["org data", "predict line"])
```

```python
import pandas as pd
import numpy as np

# Seed for reproducibility
np.random.seed(42)

# Generate 500 rows of synthetic data
age = np.random.randint(21, 60, size=500)
experience = np.clip(age - np.random.randint(18, 25, size=500), 0, None)
salary = (experience * 4500) + (age * 300) + np.random.normal(0, 5000, size=500)

# Create DataFrame
df = pd.DataFrame({
    "Age": age,
    "Experience": experience,
    "Salary": salary.round(2)
})

# Save to CSV
csv_path = "age_experience_salary.csv"
df.to_csv(csv_path, index=False)

csv_path

```
