```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
sns.scatterplot(x="age", y="fare", data=df, palette=["red", "green", "black"], hue="class") 
```

```python
sns.histplot(df["age"], bins=15, kde=True)
```

```python
sns.barplot(x=df["sex"], y=df["fare"], palette=["red", "green"])
```

```python
sns.boxplot(x=df["pclass"], y=df["fare"], palette=["red", "green"], hue=df["sex"])
```
